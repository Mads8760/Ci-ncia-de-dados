```python
#Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Gerar dados com comportamentos linear forte
np.random.seed(0) #para reproduzir os resultados
tamanho_imovel = np.random.uniform(50, 120, 80) #tamanho do metro quadrado
num_quartos = np.random.uniform(2,4,80) #numero de quartos

#Definir os coeficientes para uma relação linear forte
coef_tamanho = 1000 #Coeficiente para o tamanho do imóvel
coef_quartos = 5000 #Coeficiente para o número de quartos
intercept = 20000 #Intercept

#Gerar o valor do imoóve com ruído minimo
valor_imovel = intercept + coef_tamanho + tamanho_imovel + coef_quartos + num_quartos + np.random.normal(0,1000,80) #Valor com comportamento linear

#Criar o dataframe
df = pd.DataFrame({'tamanho_imovel':tamanho_imovel, 'num_quartos':num_quartos, 'valor_imovel':valor_imovel})
df.head()
```

```python
#Separando os dados de X e y
X = df[['tamanho_imovel', 'num_quartos']]
y = df['valor_imovel']
```

```python
#Criando uma regressão linear
model = LinearRegression()
model.fit(X,y)
```

```python
#Previsão do valor do imóvel com base nos dados de tamanho e numero de quartos
previsão = model.predict(X)
previsão
```

```python
#Visualizar resultado
plt.figure(figsize=(8,4))
plt.scatter(df['tamanho_imovel'], df['valor_imovel'], color='b', label='Dados Reais')

#Trancar a reta de regressão
z = np.polyfit(df['tamanho_imovel'], previsão, 1)
p = np.poly1d(z)
plt.plot(df['tamanho_imovel'], p(df['tamanho_imovel']), "r--", label='Regressão Linear')

plt.xlabel('Tamanho do Imóvel (m²)')
plt.ylabel('Valor do Imóvel')
plt.title('Regressão Linear para Previsão de Valores de Imóveis')
plt.legend()
plt.show()
```
